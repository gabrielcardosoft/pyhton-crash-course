# Functions

### Defining a Function

```python
def greet_user(): # Function definition: it informs Python that you're defining a function. The parentheses hold that information.
    """Display a simple greeting.""" # Docstring: it describes what the functions does.
    print("Hey, you! :¬)")

greet_user()
```

### Passing Information to a Function

```python
def greet_user(username): 
    """Display a simple greeting."""
    print(f"Hey, {username.capitalize()}! :¬)")

greet_user("gabriel")
```
